# کدام برند‌ها، برند محبوب ایرانی‌هاست؟ محبوبیت یک برند با قیمت آن چه رابطه‌ای دارد؟

```python
df5 = pd.read_excel('5.xlsx')
```

```python
df2 = pd.read_excel('2.xlsx')
```

```python
df5['product_id'] = df5['id']
```

```python
df5.head()
```

```python
df = pd.merge(df2, df5, on='product_id')
```

```python
df.head()
```

```python
brands = df['brand_name_fa'].unique()
```

```python
dict = {}
for i in range(len(brands)):
    dict[brands[i]] = len(df[df['brand_name_fa'] == brands[i]])
```

```python
import operator
sorted_brands = sorted(dict.items(), key=operator.itemgetter(1), reverse=True)
```

```python
sorted_brands
```
