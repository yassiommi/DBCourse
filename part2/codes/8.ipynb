```python
df2 = pd.read_excel('2.xlsx')
```

```python
df2.head()
```

```python
users = df2['user_id'].unique()
```

```python
df2.count()
```

```python
len(df2[df2['comment'] == 'NaN'])
```

```python
dict = {}
for i in range(len(users)):
    dict[users[i]] = len(df2[df2['user_id'] == users[i]])
```

```python
import operator
sorted_users = sorted(dict.items(), key=operator.itemgetter(1), reverse=True)
```
