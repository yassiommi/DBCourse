# در مورد مشتریان شهر‌های مختلف چه می‌توان گفت؟ مثلا تعداد مشتریان هر شهر به نسبت جمعیت آن شهر در کدام شهر‌ها بیشتر/کمتر است؟ یا کدام شهر‌ها خرید‌های گران‌تر و کدام شهر‌ها خریدهای ارزان‌تر انجام می‌دهند؟ 

```python
df3 = pd.read_csv('3.csv')
```

```python
df3.head()
```

```python
cities = df3['city_name_fa'].unique()
print(cities)
```

```python
population = pd.read_excel('Amar.xlsx')
population['pop'] = population['جمعیت']
population.head()
```

```python
df = df3
# for i in range(len(cities)):
#     df['sum of kharidars'] = df[df['city_name_fa'] == cities[i]]['ID_Customer'].unique()
```

```python
df.head()
```

```python
df = df.groupby('city_name_fa')['ID_Customer'].nunique().
```

```python
df = df.reset_index()
```

```python
df = df.rename(index=str, columns={'city_name_fa': 'city', 'ID_Customer': 'num of kharidars'})
```

```python
# df3 = df.groupby(['city_name_fa']).city_name_fa.agg('count').to_frame('Sum of Kharidars').reset_index()
```

```python
df.head()
```

```python
result = pd.merge(df, population, on='city')
```

```python
result.head()
```

```python
result['fraction'] = result['num of kharidars']/result['pop']
```

```python
result.head()
```

```python
result.sort_values(by=['fraction'], ascending=False)
```
